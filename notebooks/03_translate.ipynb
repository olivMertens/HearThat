# 03 — Translate summaries locally

Translate the chapter summaries written by notebook 01 into another language
via Azure AI **Translator Text** (passwordless) — fully local, no Blob Storage.

Set `AZURE_TRANSLATOR_ENDPOINT` in `.env` to your Translator resource endpoint,
for example `https://api.cognitive.microsofttranslator.com` (global) or your
regional endpoint.

```python
from pathlib import Path

from hearthat.translate import translate_markdown_file

src_dir = Path('../data/out/summaries')
dst_dir = Path('../data/out/summaries.es')

translated: list[Path] = []
for md in sorted(src_dir.glob('*.md')):
    out = dst_dir / md.name
    await translate_markdown_file(md, out, target_language='es')
    translated.append(out)

translated
```
