# 02 — Text to Speech

Compare three TTS backends on the same text:
1. **MAI-Voice-1** (preview, narration) via Batch Synthesis
2. **DragonHD / Omni** via Batch Synthesis
3. **gpt-4o-mini-tts** with natural-language `instructions`

```python
from pathlib import Path

from hearthat.models import VoiceSpec
from hearthat.tts import synthesize, synthesize_openai_tts

out = Path('../data/out/notebook02'); out.mkdir(parents=True, exist_ok=True)
text = 'HearThat turns books into voices. Listen closely — the narrator just woke up.'
```

## MAI-Voice-1 (Iris)

```python
mai = VoiceSpec(backend='azure_speech_mai', voice_name='en-us-Iris:MAI-Voice-1')
mai_zip = await synthesize(text, voice=mai, output_dir=out / 'mai')
mai_zip
```

## DragonHD / Omni

```python
hd = VoiceSpec(backend='azure_speech_hd', voice_name='en-US-Ava:DragonHDOmniLatestNeural')
hd_zip = await synthesize(text, voice=hd, output_dir=out / 'hd')
hd_zip
```

## gpt-4o-mini-tts (prompt-steerable)

```python
mp3 = await synthesize_openai_tts(
    text,
    output_path=out / 'openai.mp3',
    instructions='warm storyteller, slight pauses for suspense',
    voice='sage',
)
mp3
```
