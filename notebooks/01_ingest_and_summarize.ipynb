# 01 — Ingest & Summarise

Extract a PDF into chapters and summarise each one with `gpt-5.4-mini`.
All Azure calls are passwordless via `DefaultAzureCredential`.

```python
from pathlib import Path

from hearthat.config import get_settings
from hearthat.ingest import ingest_pdf
from hearthat.summarize import summarise_book

settings = get_settings()
settings.is_openai_configured, settings.is_docintel_configured
```

```python
pdf = next(Path('../data/samples').rglob('*.pdf'))
book = await ingest_pdf(pdf)
print(book.title, len(book.chapters), 'chapters')
```

```python
cost = await summarise_book(book)
print(f'cost: ${cost.summary_usd:.4f}')
print(book.chapters[0].summary)
```
